```python
# có xử lý mask
import json
import os
import cv2
import numpy as np
import pandas as pd
from tqdm import tqdm
from typing import Tuple, Optional, Dict, List

def resize_with_pad(
    image: np.ndarray,
    target_size: Tuple[int, int],
    pad_value: int = 0,
    is_mask: bool = False
) -> Tuple[np.ndarray, Dict]:
    """
    Resize ảnh giữ tỷ lệ + padding
    """
    h, w = image.shape[:2]
    target_w, target_h = target_size
    
    scale = min(target_w / w, target_h / h)
    new_w = int(w * scale)
    new_h = int(h * scale)
    
    if is_mask:
        interpolation = cv2.INTER_NEAREST
    else:
        interpolation = cv2.INTER_LINEAR
    
    resized = cv2.resize(image, (new_w, new_h), interpolation=interpolation)
    
    if len(image.shape) == 2:
        padded = np.full((target_h, target_w), pad_value, dtype=np.uint8)
    else:
        padded = np.full((target_h, target_w, image.shape[2]), pad_value, dtype=np.uint8)
    
    y_offset = (target_h - new_h) // 2
    x_offset = (target_w - new_w) // 2
    
    if len(image.shape) == 2:
        padded[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = resized
    else:
        padded[y_offset:y_offset+new_h, x_offset:x_offset+new_w, :] = resized
    
    metadata = {
        'scale': scale,
        'x_offset': x_offset,
        'y_offset': y_offset,
        'orig_size': (w, h),
        'new_size': (new_w, new_h),
        'target_size': target_size
    }
    
    return padded, metadata


def resize_no_pad(
    image: np.ndarray,
    target_size: Tuple[int, int],
    is_mask: bool = False
) -> Tuple[np.ndarray, Dict]:
    """
    Resize ảnh trực tiếp không giữ tỷ lệ (có thể bị méo)
    """
    h, w = image.shape[:2]
    target_w, target_h = target_size
    
    if is_mask:
        interpolation = cv2.INTER_NEAREST
    else:
        interpolation = cv2.INTER_LINEAR
    
    resized = cv2.resize(image, target_size, interpolation=interpolation)
    
    metadata = {
        'scale': 1.0,
        'x_offset': 0,
        'y_offset': 0,
        'orig_size': (w, h),
        'new_size': target_size,
        'target_size': target_size
    }
    
    return resized, metadata


def preprocess_image_for_both_tasks(
    image: np.ndarray,
    median_kernel: int = 5,
    resize_for_feature: Tuple[int, int] = (224, 224),
    resize_for_segmentation: Tuple[int, int] = (512, 512),
    normalize: bool = True,
    clahe: bool = True,
    clahe_clip_limit: float = 2.0,
    clahe_grid_size: Tuple[int, int] = (8, 8),
    use_padding: bool = True
) -> Dict[str, np.ndarray]:
    """
    Preprocess ảnh cho cả feature extraction và segmentation
    """
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image.copy()
    
    if median_kernel > 1:
        if median_kernel % 2 == 0:
            median_kernel += 1
        gray = cv2.medianBlur(gray, median_kernel)
    
    if clahe:
        clahe_obj = cv2.createCLAHE(clipLimit=clahe_clip_limit, tileGridSize=clahe_grid_size)
        gray = clahe_obj.apply(gray)
    
    if use_padding:
        img_feature, feature_meta = resize_with_pad(gray, resize_for_feature, is_mask=False)
        img_segmentation, seg_meta = resize_with_pad(gray, resize_for_segmentation, is_mask=False)
    else:
        img_feature, feature_meta = resize_no_pad(gray, resize_for_feature, is_mask=False)
        img_segmentation, seg_meta = resize_no_pad(gray, resize_for_segmentation, is_mask=False)
    
    if normalize:
        if img_feature.dtype != np.uint8:
            img_feature = np.uint8(np.clip(img_feature, 0, 255))
        if img_segmentation.dtype != np.uint8:
            img_segmentation = np.uint8(np.clip(img_segmentation, 0, 255))
    
    return {
        'for_feature': img_feature,
        'for_segmentation': img_segmentation,
        'feature_metadata': feature_meta,
        'segmentation_metadata': seg_meta
    }


def preprocess_mask_for_segmentation(
    mask: np.ndarray,
    target_size: Tuple[int, int],
    use_padding: bool = True
) -> Tuple[np.ndarray, Dict]:
    """
    Preprocess mask groundtruth để cùng kích thước với segmentation input
    """
    if use_padding:
        resized_mask, metadata = resize_with_pad(mask, target_size, pad_value=0, is_mask=True)
    else:
        resized_mask, metadata = resize_no_pad(mask, target_size, is_mask=True)
    
    return resized_mask, metadata


def process_dataframe_with_preprocessing(
    df: pd.DataFrame,
    image_folder: str,
    mask_folder: str,
    output_root: str,
    filename_col: str = 'FileName',
    label_col: Optional[str] = None,
    median_kernel: int = 5,
    resize_for_feature: Tuple[int, int] = (224, 224),
    resize_for_segmentation: Tuple[int, int] = (512, 512),
    clahe: bool = True,
    save_subfolder_by_label: bool = False,
    subset_name: str = 'dataset',
    use_padding: bool = True,
    auto_create_mask: bool = True  # <-- THÊM: tự tạo mask cho ảnh không có mask
) -> pd.DataFrame:
    """
    Xử lý DataFrame, tiền xử lý ảnh và mask, lưu kết quả cùng metadata
    
    Parameters:
    -----------
    auto_create_mask : bool
        Nếu True: tự tạo mask toàn 0 cho ảnh không có mask groundtruth
        Nếu False: bỏ qua ảnh không có mask
    """
    df = df.copy()
    
    # Tạo thư mục output
    feature_output_dir = os.path.join(output_root, 'feature', subset_name)
    seg_output_dir = os.path.join(output_root, 'segmentation', subset_name)
    mask_output_dir = os.path.join(output_root, 'mask', subset_name)
    metadata_dir = os.path.join(output_root, 'metadata', subset_name)
    
    os.makedirs(feature_output_dir, exist_ok=True)
    os.makedirs(seg_output_dir, exist_ok=True)
    os.makedirs(mask_output_dir, exist_ok=True)
    os.makedirs(metadata_dir, exist_ok=True)
    
    # Nếu phân loại theo label, tạo thư mục con
    if save_subfolder_by_label and label_col is not None and label_col in df.columns:
        labels = df[label_col].unique()
        for label in labels:
            os.makedirs(os.path.join(feature_output_dir, str(label)), exist_ok=True)
            os.makedirs(os.path.join(seg_output_dir, str(label)), exist_ok=True)
            os.makedirs(os.path.join(mask_output_dir, str(label)), exist_ok=True)
    
    # Thêm cột metadata vào DataFrame
    df['feature_path'] = ""
    df['segmentation_path'] = ""
    df['mask_path'] = ""
    df['has_groundtruth_mask'] = False  # <-- THÊM: đánh dấu có mask thật không
    df['orig_width'] = 0
    df['orig_height'] = 0
    df['feature_scale'] = 0.0
    df['feature_x_offset'] = 0
    df['feature_y_offset'] = 0
    df['seg_scale'] = 0.0
    df['seg_x_offset'] = 0
    df['seg_y_offset'] = 0
    df['process_status'] = ""
    
    # Duyệt từng dòng
    for idx, row in tqdm(df.iterrows(), total=len(df), desc=f"Processing {subset_name}"):
        img_name = row[filename_col]
        label = str(row[label_col]) if label_col and label_col in df.columns else 'unknown'
        
        # ==================== XỬ LÝ ẢNH GỐC ====================
        name, ext = os.path.splitext(img_name)
        img_path = None
        possible_image_extensions = ['.jpg', '.jpeg', '.png', '.PNG', '.JPG', '.JPEG']
        for try_ext in possible_image_extensions:
            test_path = os.path.join(image_folder, f"{name}{try_ext}")
            if os.path.exists(test_path):
                img_path = test_path
                break

        if img_path is None:
            df.at[idx, 'process_status'] = f"Missing image: {name}"
            continue
        
        img = cv2.imread(img_path)
        if img is None:
            df.at[idx, 'process_status'] = f"Cannot read image: {img_path}"
            continue
        
        # ==================== XỬ LÝ MASK ====================        
        # Thử tìm mask thực tế
        name, ext = os.path.splitext(img_name)
        
        mask_path = None
        possible_extensions = ['.png', '.PNG', '.tif', '.TIF']  # Đuôi của mask
        for try_ext in possible_extensions:
            test_path = os.path.join(mask_folder, 'defect', f"{name}{try_ext}")
            if os.path.exists(test_path):
                mask_path = test_path
                break
        
        has_gt_mask = mask_path is not None
        df.at[idx, 'has_groundtruth_mask'] = has_gt_mask
        
        # Nếu không có mask và không cho phép tự tạo -> bỏ qua
        if not has_gt_mask and not auto_create_mask:
            df.at[idx, 'process_status'] = f"Skipped: no mask for {name}"
            continue
        
        # Đọc mask nếu có, nếu không thì tạo mask toàn 0
        if has_gt_mask:
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is None:
                df.at[idx, 'process_status'] = f"Cannot read mask: {mask_path}"
                continue
        else:
            # Tạo mask toàn 0 (background) với kích thước bằng ảnh gốc
            h, w = img.shape[:2]
            mask = np.zeros((h, w), dtype=np.uint8)
        
        try:
            # ==================== PREPROCESS ẢNH ====================
            processed = preprocess_image_for_both_tasks(
                img,
                median_kernel=median_kernel,
                resize_for_feature=resize_for_feature,
                resize_for_segmentation=resize_for_segmentation,
                clahe=clahe,
                normalize=True,
                use_padding=use_padding
            )
            
            # ==================== PREPROCESS MASK ====================
            mask_resized, mask_meta = preprocess_mask_for_segmentation(
                mask,
                target_size=resize_for_segmentation,
                use_padding=use_padding
            )
            
            # Tên file output
            output_filename = f"{name}.png"
            
            # Xác định thư mục output
            if save_subfolder_by_label and label_col is not None:
                feature_save_dir = os.path.join(feature_output_dir, label)
                seg_save_dir = os.path.join(seg_output_dir, label)
                mask_save_dir = os.path.join(mask_output_dir, label)
            else:
                feature_save_dir = feature_output_dir
                seg_save_dir = seg_output_dir
                mask_save_dir = mask_output_dir
            
            # Đường dẫn đầy đủ
            feature_path = os.path.join(feature_save_dir, output_filename)
            seg_path = os.path.join(seg_save_dir, output_filename)
            mask_path_saved = os.path.join(mask_save_dir, output_filename)
            
            # Lưu ảnh
            cv2.imwrite(feature_path, processed['for_feature'])
            cv2.imwrite(seg_path, processed['for_segmentation'])
            cv2.imwrite(mask_path_saved, mask_resized)
            
            # Lấy metadata
            seg_meta = processed['segmentation_metadata']
            feature_meta = processed['feature_metadata']
            
            # Lưu metadata vào DataFrame
            df.at[idx, 'feature_path'] = feature_path
            df.at[idx, 'segmentation_path'] = seg_path
            df.at[idx, 'mask_path'] = mask_path_saved
            df.at[idx, 'orig_width'] = seg_meta['orig_size'][0]
            df.at[idx, 'orig_height'] = seg_meta['orig_size'][1]
            df.at[idx, 'seg_scale'] = seg_meta['scale']
            df.at[idx, 'seg_x_offset'] = seg_meta['x_offset']
            df.at[idx, 'seg_y_offset'] = seg_meta['y_offset']
            df.at[idx, 'feature_scale'] = feature_meta['scale']
            df.at[idx, 'feature_x_offset'] = feature_meta['x_offset']
            df.at[idx, 'feature_y_offset'] = feature_meta['y_offset']
            df.at[idx, 'process_status'] = 'success' if has_gt_mask else 'success_auto_mask'
            
            # Lưu metadata riêng cho ảnh dưới dạng JSON
            json_metadata = {
                'image_id': name,
                'subset': subset_name,
                'label': label,
                'has_groundtruth_mask': has_gt_mask,
                'original': {
                    'width': seg_meta['orig_size'][0],
                    'height': seg_meta['orig_size'][1]
                },
                'feature': {
                    'path': feature_path,
                    'target_size': resize_for_feature,
                    'scale': feature_meta['scale'],
                    'x_offset': feature_meta['x_offset'],
                    'y_offset': feature_meta['y_offset']
                },
                'segmentation': {
                    'path': seg_path,
                    'target_size': resize_for_segmentation,
                    'scale': seg_meta['scale'],
                    'x_offset': seg_meta['x_offset'],
                    'y_offset': seg_meta['y_offset']
                },
                'mask': {
                    'path': mask_path_saved,
                    'target_size': resize_for_segmentation,
                    'scale': mask_meta['scale'],
                    'x_offset': mask_meta['x_offset'],
                    'y_offset': mask_meta['y_offset'],
                    'is_auto_generated': not has_gt_mask  # <-- Đánh dấu mask tự tạo
                },
                'mask_threshold': 0.5
            }
            
            json_path = os.path.join(metadata_dir, f"{name}.json")
            with open(json_path, 'w') as f:
                json.dump(json_metadata, f, indent=2)
            
        except Exception as e:
            df.at[idx, 'process_status'] = f"Error: {str(e)}"
    
    # Thống kê
    success_count = df['process_status'].str.contains('success').sum()
    auto_mask_count = (df['process_status'] == 'success_auto_mask').sum()
    gt_mask_count = (df['process_status'] == 'success').sum()
    
    print(f"\n[INFO] Hoàn thành: {success_count}/{len(df)} ảnh được xử lý thành công")
    print(f"  - Có mask groundtruth: {gt_mask_count} ảnh")
    print(f"  - Tự tạo mask (ảnh normal): {auto_mask_count} ảnh")
    
    return df


def process_multiple_dataframes(
    dataframes: Dict[str, pd.DataFrame],
    image_root: str,
    mask_root: str,
    output_root: str,
    filename_col: str = 'FileName',
    label_col: Optional[str] = None,
    median_kernel: int = 5,
    resize_for_feature: Tuple[int, int] = (224, 224),
    resize_for_segmentation: Tuple[int, int] = (512, 512),
    clahe: bool = True,
    save_subfolder_by_label: bool = False,
    use_padding: bool = True,
    auto_create_mask: bool = True  # <-- THÊM THAM SỐ
) -> Dict[str, pd.DataFrame]:
    """
    Xử lý nhiều DataFrame cùng lúc
    """
    results = {}
    
    for subset_name, df in dataframes.items():
        image_folder = os.path.join(image_root, subset_name)
        mask_folder = os.path.join(mask_root, subset_name)
        
        if not os.path.exists(image_folder):
            image_folder = image_root
        if not os.path.exists(mask_folder):
            mask_folder = mask_root
        
        processed_df = process_dataframe_with_preprocessing(
            df=df,
            image_folder=image_folder,
            mask_folder=mask_folder,
            output_root=output_root,
            filename_col=filename_col,
            label_col=label_col,
            median_kernel=median_kernel,
            resize_for_feature=resize_for_feature,
            resize_for_segmentation=resize_for_segmentation,
            clahe=clahe,
            save_subfolder_by_label=save_subfolder_by_label,
            subset_name=subset_name,
            use_padding=use_padding,
            auto_create_mask=auto_create_mask  # <-- TRUYỀN THAM SỐ
        )
        results[subset_name] = processed_df
        
        # Lưu DataFrame
        os.makedirs(os.path.join(output_root, 'CSV'), exist_ok=True)
        output_csv = os.path.join(output_root, 'CSV', f"{subset_name}_processed.csv")
        processed_df.to_csv(output_csv, index=False)
        print(f"[INFO] Đã lưu CSV: {output_csv}")
    
    return results
```

```python
import os
import shutil
import cv2
import numpy as np
import pandas as pd
from pathlib import Path
from tqdm import tqdm
from typing import Dict, List, Optional, Tuple
from datetime import datetime


class RawImageExtractor:
    """
    Trích xuất ảnh raw từ các DataFrame và lưu vào thư mục mới
    Sử dụng logic tương tự process_dataframe_with_preprocessing
    """
    
    def __init__(self, 
                 dataframes: Dict[str, pd.DataFrame],
                 image_root: str,
                 output_dir: str,
                 filename_col: str = 'FileName',
                 label_col: Optional[str] = None,
                 save_subfolder_by_label: bool = False,
                 copy_images: bool = True,
                 max_images_per_subset: Optional[int] = None):
        """
        Khởi tạo bộ trích xuất ảnh raw
        
        Args:
            dataframes: Dictionary {subset_name: DataFrame}
            image_root: Thư mục gốc chứa các folder con (T1_S148_I108_1, ...)
            output_dir: Thư mục đích để lưu ảnh đã trích xuất
            filename_col: Tên cột chứa tên file ảnh
            label_col: Tên cột chứa nhãn (nếu có)
            save_subfolder_by_label: Có phân loại theo label không
            copy_images: Copy ảnh (True) hay tạo symlink (False)
            max_images_per_subset: Số lượng ảnh tối đa mỗi subset
        """
        self.dataframes = dataframes
        self.image_root = Path(image_root)
        self.output_dir = Path(output_dir)
        self.filename_col = filename_col
        self.label_col = label_col
        self.save_subfolder_by_label = save_subfolder_by_label
        self.copy_images = copy_images
        self.max_images_per_subset = max_images_per_subset
        
        # Tạo thư mục output
        self.output_dir.mkdir(parents=True, exist_ok=True)
        
        # Thống kê
        self.stats = {
            'total_found': 0,
            'total_copied': 0,
            'total_missing': 0,
            'per_subset': {}
        }
    
    def _find_image_path(self, img_name: str, image_folder: Path) -> Optional[Path]:
        """
        Tìm đường dẫn đầy đủ của ảnh (giống logic trong preprocess)
        
        Args:
            img_name: Tên file ảnh (có thể có hoặc không có đuôi mở rộng)
            image_folder: Thư mục chứa ảnh
        """
        # Tách tên và đuôi mở rộng
        name, ext = os.path.splitext(img_name)
        
        # Các đuôi mở rộng có thể có
        possible_extensions = ['.jpg', '.jpeg', '.png', '.PNG', '.JPG', '.JPEG']
        
        # Nếu đã có đuôi mở rộng trong tên, thử trực tiếp
        if ext in possible_extensions:
            test_path = image_folder / img_name
            if test_path.exists():
                return test_path
        
        # Thử thêm các đuôi mở rộng
        for try_ext in possible_extensions:
            test_path = image_folder / f"{name}{try_ext}"
            if test_path.exists():
                return test_path
        
        return None
    
    def extract_images(self) -> Dict:
        """
        Trích xuất tất cả ảnh từ các DataFrame
        
        Returns:
            Dict: Thống kê kết quả
        """
        print("=" * 70)
        print("🚀 BẮT ĐẦU TRÍCH XUẤT ẢNH RAW")
        print("=" * 70)
        print(f"📁 Thư mục gốc: {self.image_root}")
        print(f"📁 Thư mục đích: {self.output_dir}")
        print(f"📋 Cột tên file: {self.filename_col}")
        print(f"🏷️  Phân loại theo label: {self.save_subfolder_by_label}")
        print(f"📦 Copy ảnh: {self.copy_images}")
        print("=" * 70)
        
        for subset_name, df in self.dataframes.items():
            print(f"\n📊 Đang xử lý subset: {subset_name}")
            
            # Kiểm tra cột tồn tại
            if self.filename_col not in df.columns:
                print(f"   ⚠️ Không tìm thấy cột '{self.filename_col}', bỏ qua")
                continue
            
            # Lấy danh sách tên file
            image_names = df[self.filename_col].dropna().astype(str).tolist()
            image_names = [name.strip() for name in image_names if name.strip()]
            
            # Giới hạn số lượng nếu cần
            if self.max_images_per_subset:
                image_names = image_names[:self.max_images_per_subset]
            
            # Xác định thư mục ảnh nguồn
            image_folder = self.image_root / subset_name
            if not image_folder.exists():
                # Thử dùng image_root trực tiếp
                image_folder = self.image_root
                if not image_folder.exists():
                    print(f"   ❌ Thư mục không tồn tại: {image_folder}")
                    continue
            
            print(f"   📁 Thư mục nguồn: {image_folder}")
            print(f"   📸 Số ảnh cần xử lý: {len(image_names)}")
            
            # Thống kê subset
            subset_found = 0
            subset_copied = 0
            subset_missing = 0
            
            # Tạo thư mục đích cho subset này
            subset_output_dir = self.output_dir / subset_name
            subset_output_dir.mkdir(exist_ok=True)
            
            # Tạo thư mục con theo label nếu cần
            label_dirs = {}
            if self.save_subfolder_by_label and self.label_col and self.label_col in df.columns:
                # Tạo mapping từ tên ảnh đến label
                name_to_label = dict(zip(df[self.filename_col], df[self.label_col]))
            
            for img_name in tqdm(image_names, desc=f"   Đang xử lý {subset_name}"):
                img_path = self._find_image_path(img_name, image_folder)
                
                if img_path and img_path.exists():
                    subset_found += 1
                    
                    # Xác định thư mục đích
                    if self.save_subfolder_by_label and self.label_col and self.label_col in df.columns:
                        label = str(name_to_label.get(img_name, 'unknown'))
                        dest_folder = subset_output_dir / label
                        dest_folder.mkdir(exist_ok=True)
                    else:
                        dest_folder = subset_output_dir
                    
                    dest_path = dest_folder / img_path.name
                    
                    # Copy hoặc tạo symlink
                    try:
                        if self.copy_images:
                            shutil.copy2(img_path, dest_path)
                        else:
                            # Tạo symlink (cần quyền admin trên Windows)
                            if os.name == 'nt':  # Windows
                                import subprocess
                                subprocess.run(['mklink', str(dest_path), str(img_path)], shell=True)
                            else:  # Linux/Mac
                                os.symlink(img_path, dest_path)
                        subset_copied += 1
                    except Exception as e:
                        print(f"      ⚠️ Lỗi khi copy {img_path.name}: {e}")
                        subset_missing += 1
                else:
                    subset_missing += 1
                    print(f"      ⚠️ Không tìm thấy ảnh: {img_name}")
            
            # Lưu thống kê
            self.stats['per_subset'][subset_name] = {
                'total': len(image_names),
                'found': subset_found,
                'copied': subset_copied,
                'missing': subset_missing,
                'output_dir': str(subset_output_dir)
            }
            self.stats['total_found'] += subset_found
            self.stats['total_copied'] += subset_copied
            self.stats['total_missing'] += subset_missing
            
            print(f"   ✅ Tìm thấy: {subset_found}/{len(image_names)}")
            print(f"   📁 Đã lưu vào: {subset_output_dir}")
        
        # Tổng kết
        self._print_summary()
        self._save_summary()
        
        return self.stats
    
    def _print_summary(self):
        """In báo cáo tổng kết"""
        print("\n" + "=" * 70)
        print("📊 TỔNG KẾT TRÍCH XUẤT")
        print("=" * 70)
        print(f"📸 Tổng số ảnh tìm thấy: {self.stats['total_found']}")
        print(f"✅ Tổng số ảnh đã copy: {self.stats['total_copied']}")
        print(f"❌ Tổng số ảnh thiếu: {self.stats['total_missing']}")
        
        print("\n📁 Chi tiết theo subset:")
        for subset, stats in self.stats['per_subset'].items():
            print(f"   - {subset}: {stats['copied']}/{stats['total']} ảnh")
        
        print(f"\n📂 Thư mục output: {self.output_dir}")
        print("=" * 70)
    
    def _save_summary(self):
        """Lưu báo cáo vào file"""
        summary_path = self.output_dir / "extraction_summary.json"
        
        summary = {
            'timestamp': datetime.now().isoformat(),
            'image_root': str(self.image_root),
            'output_dir': str(self.output_dir),
            'copy_images': self.copy_images,
            'save_subfolder_by_label': self.save_subfolder_by_label,
            'stats': self.stats
        }
        
        import json
        with open(summary_path, 'w', encoding='utf-8') as f:
            json.dump(summary, f, indent=2, ensure_ascii=False)
        
        print(f"\n💾 Đã lưu báo cáo tại: {summary_path}")
    
    def create_metadata_csv(self):
        """Tạo file CSV metadata cho ảnh đã trích xuất"""
        all_records = []
        
        for subset_name, stats in self.stats['per_subset'].items():
            subset_dir = Path(stats['output_dir'])
            
            # Duyệt tìm tất cả ảnh
            for ext in ['*.jpg', '*.jpeg', '*.png']:
                for img_path in subset_dir.rglob(ext):
                    # Xác định label từ đường dẫn
                    if self.save_subfolder_by_label:
                        label = img_path.parent.name
                    else:
                        label = 'unknown'
                    
                    # Lấy kích thước ảnh
                    try:
                        img = cv2.imread(str(img_path))
                        height, width = img.shape[:2] if img is not None else (0, 0)
                    except:
                        height, width = 0, 0
                    
                    all_records.append({
                        'subset': subset_name,
                        'image_name': img_path.name,
                        'image_path': str(img_path),
                        'label': label,
                        'width': width,
                        'height': height
                    })
        
        if all_records:
            df = pd.DataFrame(all_records)
            csv_path = self.output_dir / "extracted_images_metadata.csv"
            df.to_csv(csv_path, index=False)
            print(f"💾 Đã tạo file metadata CSV tại: {csv_path}")
            return df
        return None


def extract_images_for_batches(
    dataframes: Dict[str, pd.DataFrame],
    image_root: str,
    output_dir: str,
    batch_size: int = 500,
    filename_col: str = 'FileName'
) -> Dict:
    """
    Trích xuất ảnh raw và chia thành các batch nhỏ
    
    Args:
        dataframes: Dictionary {subset_name: DataFrame}
        image_root: Thư mục gốc chứa ảnh
        output_dir: Thư mục output cho batch
        batch_size: Số lượng ảnh mỗi batch
        filename_col: Tên cột chứa tên file
    """
    # Bước 1: Trích xuất tất cả ảnh vào thư mục tạm
    extractor = RawImageExtractor(
        dataframes=dataframes,
        image_root=image_root,
        output_dir=Path(output_dir) / "extracted_raw",
        filename_col=filename_col,
        copy_images=True
    )
    
    extractor.extract_images()
    
    # Bước 2: Chia thành các batch
    print("\n" + "=" * 70)
    print("🚀 CHIA ẢNH THÀNH CÁC BATCH")
    print("=" * 70)
    
    # Lấy tất cả ảnh đã trích xuất
    extracted_dir = Path(output_dir) / "extracted_raw"
    all_images = []
    for ext in ['*.jpg', '*.jpeg', '*.png']:
        all_images.extend(extracted_dir.rglob(ext))
    
    print(f"📸 Tổng số ảnh đã trích xuất: {len(all_images)}")
    
    # Tạo thư mục batch
    batches_dir = Path(output_dir) / "batches"
    batches_dir.mkdir(parents=True, exist_ok=True)
    
    # Chia batch
    num_batches = (len(all_images) + batch_size - 1) // batch_size
    print(f"📦 Chia thành {num_batches} batch, mỗi batch {batch_size} ảnh")
    
    batch_info = []
    
    for batch_idx in range(num_batches):
        start = batch_idx * batch_size
        end = min(start + batch_size, len(all_images))
        batch_images = all_images[start:end]
        
        batch_folder = batches_dir / f"batch_{batch_idx + 1:03d}"
        batch_folder.mkdir(exist_ok=True)
        
        # Copy ảnh vào batch folder
        for img_path in tqdm(batch_images, desc=f"Batch {batch_idx + 1}"):
            # Giữ nguyên cấu trúc thư mục con
            rel_path = img_path.relative_to(extracted_dir)
            dest_path = batch_folder / rel_path
            dest_path.parent.mkdir(parents=True, exist_ok=True)
            shutil.copy2(img_path, dest_path)
        
        batch_info.append({
            'batch_id': batch_idx + 1,
            'num_images': len(batch_images),
            'folder': str(batch_folder),
            'images': [str(p.relative_to(batch_folder)) for p in batch_images]
        })
        
        print(f"   Batch {batch_idx + 1}: {len(batch_images)} ảnh -> {batch_folder}")
    
    # Lưu thông tin batch
    info_path = batches_dir / "batch_info.json"
    import json
    with open(info_path, 'w', encoding='utf-8') as f:
        json.dump(batch_info, f, indent=2, ensure_ascii=False)
    
    print(f"\n✅ Đã lưu thông tin batch tại: {info_path}")
    print(f"📁 Thư mục batches: {batches_dir}")
    
    return {
        'batches_dir': batches_dir,
        'batch_info': batch_info,
        'total_images': len(all_images),
        'num_batches': num_batches
    }


# ========== SỬ DỤNG ==========

if __name__ == "__main__":
    # Giả sử bạn đã có dictionary các DataFrame
    # dfs = {
    #     'T1_S148_I108_1': t1_s148_i108_1,
    #     'T1_S148_I108_2': t1_s148_i108_2,
    #     ...
    # }
    
    # Đường dẫn thư mục gốc chứa ảnh raw
    IMAGE_ROOT = r"D:\Study\On-going\Xu_ly_anh\Computer Vision\defect_detection\data\raw\TSfabrics\IMAGE"
    
    # Thư mục output
    OUTPUT_DIR = r"D:\Study\On-going\Xu_ly_anh\Computer Vision\defect_detection\data\extracted_raw_images"
    
    # Cách 1: Chỉ trích xuất ảnh, không chia batch
    extractor = RawImageExtractor(
        dataframes=dfs,  # Dictionary DataFrame của bạn
        image_root=IMAGE_ROOT,
        output_dir=OUTPUT_DIR,
        filename_col='FileName',
        label_col='Label',
        save_subfolder_by_label=False,  # Phân loại theo label
        copy_images=True
    )
    
    result = extractor.extract_images()
    
    # Tạo file metadata CSV
    extractor.create_metadata_csv()
    
    # Cách 2: Trích xuất và chia batch
    # batch_result = extract_images_for_batches(
    #     dataframes=dfs,
    #     image_root=IMAGE_ROOT,
    #     output_dir=OUTPUT_DIR,
    #     batch_size=500,
    #     filename_col='FileName'
    # )
```
